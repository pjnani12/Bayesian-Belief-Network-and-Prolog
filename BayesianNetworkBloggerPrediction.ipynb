# 1.	Construct a Bayesian Belief Network for the given data.

```python
## we use the from samples method of pomegranate model to create the Bayseian Network and the exact algorithm
```

```python
##learning the structure from the data
model = BayesianNetwork.from_samples(df.values, algorithm='exact',state_names=['Degree', 'caprice', 'topic', 'lmt', 'lpss', 'pb']
                                    ,exclude_edges=[(0,1)])
```

```python
model.structure
```

```python
model.plot('BBN.pdf')
```

```python
from IPython.display import IFrame
IFrame("BBN.pdf", width=600, height=300)## the network is also provided in the word file attached with the submission.
```

```python
##Calculating the CPT table entries for calculating the joint prorbabilities
model.fit(df.values)
```

```python
##baking the model for inference
model.bake()
```

# 2.	Predict the probability of  data having the following values : Education: high, Political caprice: left, Topics: political, Local Media Turnover: yes, Local Political and Social Space: yes, who is a professional blogger.

```python
model.probability([['high','left','political','yes','yes','yes']])
```

# 3.	Predict the class with the following attribute values: A blogger with a high education level, who has middle political caprice, who blogs about travel and tourism with no local media turnover and has local political and social space.

```python
a=model.predict([['high','middle','tourism','no','yes',None]])
print(a)
print("The prediction for pb is {0}".format(np.array(a)[0,5]))
```

# 4.	Infer the probability for the data : with values : Low education, left caprice, who blogs on news and has local media turnover and has political and social space but is not a professional blogger

```python
print(model.probability([['low','left','news','yes','yes','no']]))
```
